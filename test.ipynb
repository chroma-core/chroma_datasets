```python
# download data from hugging face
from datasets import load_dataset
dataset = load_dataset("glue", "ax")
```

```python
import chromadb
from chroma_datasets import import_dataset

client = chromadb.Client()

mapping = {
    "ids": lambda row: str(row["idx"]),
    "metadatas": lambda row: {"premise": row["premise"], "hypothesis": row["hypothesis"], "label": row["label"]},
    "documents": "premise"
}

import_dataset(chroma_client=client, collection_name="HF-dataset", dataset=dataset["test"], mapping_function=mapping, dataset_type="HuggingFaceDataset")
collection = client.get_collection("HF-dataset")
collection.get("1",include=["documents", "embeddings"])
```
